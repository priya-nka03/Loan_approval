```python
df.columns=df.columns.astype(str).str.strip().str.replace(r'\s+',"",regex=True)
```

```python
df["loan_status"]
```

```python
df['loan_status'] = (
    df['loan_status']
    .astype(str)
    .str.replace(r'\s+', '', regex=True)
)
```

```python
def status_bin(x):
    if x=="Rejected":
        return 0
    if x=="Approved":
        return 1
    else :
        return "not availabe"
df["loan_status"]=df["loan_status"].apply(status_bin)
```

```python
df["total_assets"]=df["residential_assets_value"]+df["commercial_assets_value"]+df["luxury_assets_value"]+df["bank_asset_value"]
```

```python
df=df.drop(columns="total_assest")
```
