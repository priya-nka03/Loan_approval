# Cibil Score Vs Loan Status

```python
df.groupby('loan_status')['cibil_score'].mean()
```

Applicant with higher cibil score get their loan approved and rest does not.

On an Average of 703 cibil score Applicant Loan is approved and below 430 the loan is not approved.

Approved loans have significantly higher average CIBIL scores, confirming credit history as the primary driver of loan approval decisions.

# Income-Loan Ratio vs Loan Status

```python
df["income_loan_ratio"]=df["income_annum"]/df["loan_amount"]
```

```python
df.head()
```

```python
df.describe()
```

```python
df.groupby("loan_status")["income_loan_ratio"].mean()
```

no much difference


The income-to-loan ratio shows very minimal varaition between approved and rejected loans, indicating that income alone does not strongly determine approval. It acts as a secondary validation check when primarys are done.

# Total Assets Vs Loan Status

```python
df.groupby("loan_status")["total_assets"].median()
```

The difference is only of 9lakh which is a very small difference. 

So Total assets alone are not a strong predictor for loan approval

The Median of total assets differ marginally between approved and rejected applicants, suggesting that assets ownership alone does not guarantee loan approval. It can considered as a supporting factor but not the primary one.

# Education Vs Loan Status

```python
pd.crosstab(df['education'], df['loan_status'], normalize=True)*100
```

```python
pd.crosstab(df['education'], df['loan_status'])
```

Cross Table analysis shows there is not much difference but applicants with higher qualication have relatively better approval rates, implies education may indirectly influence approvals 

```python
pd.crosstab(df["self_employed"],df["loan_status"], normalize=True)*100
```

```python
pd.crosstab(df["self_employed"],df["loan_status"])
```

Salaried applicants show relatively higher approval proportions compared to self-employed customers, reflecting banks’ preference for stable and predictable income sources rather than variable earnings.

# Final Insight

Loan approval decisions are primarily driven by credit score, with income, assets, education, employment type, and dependents acting as supporting factors rather than independent decision-makers. This suggests a credit-first risk assessment approach, where additional attributes are used to validate eligibility rather than determine approval on it.
